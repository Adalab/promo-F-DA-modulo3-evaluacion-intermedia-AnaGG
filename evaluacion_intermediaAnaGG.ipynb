3. Preparación de los datos:

Los nombres de las columnas empiezan con mayúsculas, pon todos los nombres de las columnas en minúsculas.

```python
df_spoty.columns
```

```python
nuevas_columnas = {columna: columna.lower() for columna in df_spoty.columns}
nuevas_columnas
```

```python
df_spoty.rename(columns = nuevas_columnas, inplace = True)
df_spoty.sample(20)
```

Haciendo el análisis exploratorio te deberías haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Crea una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

```python
def cambiar_comas(cadena):
    try:
        return float(cadena.replace(',','.'))
    except:
        np.nan
```

```python
df_spoty["danceability"] = df_spoty["danceability"].apply(cambiar_comas)
```

```python
df_spoty["energy"] = df_spoty["energy"].apply(cambiar_comas)
```

```python
df_spoty["key"] = df_spoty["key"].apply(cambiar_comas)
```

```python
df_spoty["loudness"] = df_spoty["loudness"].apply(cambiar_comas)
```

```python
df_spoty["speechiness"] = df_spoty["speechiness"].apply(cambiar_comas)
```

```python
df_spoty["acousticness"] = df_spoty["acousticness"].apply(cambiar_comas)
```

```python
df_spoty["instrumentalness"] = df_spoty["instrumentalness"].apply(cambiar_comas)
```

Después de haber hecho los cambios, chequea los tipos de datos. ¿Son ya del tipo correcto? En caso de que no, crea otra función o modifica la anterior para que sean de tipo float.

```python
df_spoty[['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']].dtypes
```

Hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haz lo mismo para las columnas de "title" y "album".

```python
df_spoty['title'] = df_spoty['title'].str.lower()
df_spoty['album'] = df_spoty['album'].str.lower()
df_spoty.sample(20)
```

Algunos de los artistas tienen símbolos raros, en concreto el símbolo $. Crea una función que los reemplaze pos "s".

```python
df_spoty["artist"] = df_spoty["artist"].str.replace("$", "s")
```
